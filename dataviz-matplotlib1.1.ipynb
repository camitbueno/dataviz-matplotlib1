#### Nosso objetivo agora será verificar se existem dias da semana que são mais quentes ou mais frios!

```python
# importando as bibliotecas e o dataset 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#faz com que nossos gráficos sejam exibidos automaticamente
import datetime

df = pd.read_csv('dados/monitoramento_tempo.csv')
df.head()
```

Lembrando que a nossa temperatura do dataset encontra-se em Kelvin e não em Celsius.

```python
# aqui buscaremos a média de temperatura por dias da semana
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
temperatura_por_dia_da_semana 
```

Percebemos que as temperaturad por dia de semana são bem parecidas, o que "quebra" nossa ideia de verificarmos se existem dias da semana que são mais quentes ou mais frios. 
Entretanto, continuaremos com a plotagem dos gráficos e ordenaremos os dias da semana para a ordem dos mesmos e não em ordem alfabética como o Python nos devolveu.

```python
# ordenando os dias da semana
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]
temperatura_por_dia_da_semana
```

Agora, vamos visualizar esse resultado?

### 1. Primeiras Visualizações

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
indice = range(len(temperatura_por_dia_da_semana))

eixo.bar(indice, temperatura_por_dia_da_semana)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura média (K)', fontsize=15) # lembrando que estamos trabalhando com as médias encontradas acima
```

Nosso gráfico de barras com as médias está pronto, mas podemos melhorá-lo cada vez mais. Note que os dias da semana estão enumerados e não nomeados como estamos acostumados, vamos ajustar isso agora.

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
#começando o tratamento para nomear os dias da semana
indice = range(len(temperatura_por_dia_da_semana)) 

eixo.bar(indice, temperatura_por_dia_da_semana)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura média (K)', fontsize=15)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

Pronto, agora também podemos alterar as cores das barras, deixando uma cor para cada dia da semana.

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
indice = range(len(temperatura_por_dia_da_semana))
cores = ['black', 'r', 'g', 'b', 'yellow', 'orange', 'magenta'] # colorindo as barras

eixo.bar(indice, temperatura_por_dia_da_semana, color = cores)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura média (K)', fontsize=15)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

Está ok, mas ainda podemos customizar ainda mais... Vamos adicionar bordas em nossas barras agora.

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])
indice = range(len(temperatura_por_dia_da_semana))
cores = ['black', 'r', 'g', 'b', 'yellow', 'orange', 'magenta']

eixo.bar(indice, temperatura_por_dia_da_semana, color = cores, edgecolor='black') # edgecolor define nossa borda
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura média (K)', fontsize=15)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

As informações do nosso gráfico estão bem destacadas, podemos identificar visualmente que as médias são realmente bem próximas, de fato anulando nossa ideia de verificação inicial.

### 2. ...Pizza?

Nesse caso em específico podemos utilizar um gráfico de pizza para visualizarmos nossas informações. 
Lembrando que deve-se ter muito cuidado com a utilização desse tipo de gráfico. 
Mas, para o nosso caso, olharemos para ele com carinho, pois ele mostrará que todos os dias da semana possui a mesma média.
Vamos lá?

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index)
eixo.set_title('Temperatura por dia da semana', fontsize =15, pad=10)
```

Seria interessante incluirmos no gráfico a porcentagem de cada dia da semana. Para isso, incluiremos na chamada de eixo.pie() o parâmetro autopct recebendo como valor %.1f$%% (onde .1 significa que teremos apenas uma casa decimal).

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index,
        autopct='%.1f%%')
eixo.set_title('Temperatura por dia da semana', fontsize =15, pad=10)
```

Todas as fatias possuem o valor de 14.3% pois a definição '%.1f%%' também arredonda valores.

Temos um recurso disponível que podemos "explodir" da imagem determinados valores. Vamos selecionar somente os finais de semana e visualizar isso? 

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [0.1,0,0,0,0,0.1,0.1] # aqui geramos a explosão de maneira suave, pois se não colocassemos o 0 antecedendo nossos 1
                                # teríamos gerado fatias extremamente distantes umas das outras na explosão

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index,
        autopct='%.1f%%', explode = explodir)
eixo.set_title('Temperatura por dia da semana', fontsize =15, pad=10)
```

### Conclusão

Conseguimos gerar tanto um gráfico de barras quanto um gráfico de pizza para ilustrarmos uma informação - a de que todos os dias da semana possuem uma temperatura média bastante semelhante, o que nega nossa hipótese inicial de que alguns dias poderiam ser mais quentes ou mais frios.
